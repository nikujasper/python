```python
import random
score=0
i=1
lst=["real","is","rare","fake","everywhere","software","hacker"]
lstr=["LARE","SI","ARRE","KEFA","WHEREEVERY","WARESOFT","KERHAC"]

while(i<=5):
  temp=random.choice(lstr)
  print("ROUND",i)
  str= input("Enter correct word for "+temp+"=")
  if str in lst:
    print("correct")
    print("_______")
    score+=1
      
  else:
    print("incorrect")
    print("_________")
    
  i+=1
print("Your total score=",score)
```
